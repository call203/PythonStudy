# 네이버 POST 요청(데이터랩)

```python
url ='https://openapi.naver.com/v1/datalab/search'
headers={
     'X-Naver-Client-Id': '',
    'X-Naver-Client-Secret': '',
    'Contnet-Type' :'application/json'
}

params={
    'startDate':'2019-08-03',
    "endDate": "2020-08-03",
    "timeUnit": "date",
    "keywordGroups":[
        {
            "groupName":'코로나',
            'keywords' :['코로나 현황','코로나','코로나 확산']
        },{
            "groupName":'마스크',
            'keywords' :['마스크','마스크 구매','KF94']
        }
    ]
    
}
```

#####  만약 xml로 만들어 보내야 할 경우 dicttoxml 라이브러리 설치 및 import하여 사용

```python
import json
data = json.dumps(params)
```

```python
resp = requests.post(url, headers=headers,data=data)
resp
```

```python
result = resp.json()
result
```
