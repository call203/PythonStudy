```python
ktx_df['경부선 KTX'].where(lambda x: x>=5000,'<5000')
ktx_df['경부선 KTX'].where(lambda x: x<5000,'>=5000')
```

```python
def categorizing_by_5000(data):
    data = data.where(lambda x: x>=5000,0)
    data = data.where(lambda x : x<5000,1)
   
    return data
```

```python
#inplace사용하면 실제값도 바뀜
ktx_df.iloc[:,:-2].apply(categorizing_by_5000)
```
