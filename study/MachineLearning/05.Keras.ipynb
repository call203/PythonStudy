```python
from keras.models import Sequential
from keras.layers import Dense
from keras import optimizers
import numpy as np
import tensorflow as tf
```

```python
seed =0
np.random.seed(seed)
tf.random.set_seed(seed)

Data_set = np.loadtxt('./dataset/ThoraricSurgery.csv',delimiter=",")
X = Data_set[:,0:17]
Y = Data_set[:,17]

#딥러닝 구조를 결정
model = Sequential()
model.add(Dense(10,input_dim=17, activation='relu')) # 이 층에서 노드를 몇개 만들지 결정
model.add(Dense(1,activation='sigmoid'))   #출력층

#딥러닝을 실행합니다.
model.compile(loss='mean_squared_error',optimizer='adam', metrics=['accuracy'])
model.fit(X,Y,epochs=30,batch_size=10) # ==session.run() , epochs:반복수 , batch_size:샘플 한번에 몇개 처리?


#결과 출력
print("\nAccuracy:%.4f" % model.evaluate(X,Y)[1]) # 정확도 평가
```

## XOR를 케라스로 해보기

```python
training_data = np.array([[0,0],[0,1],[1,0],[1,1]], "float32")
target_data = np.array([[0],[1],[1],[0]],"float32")

model = Sequential()
model.add(Dense(5,input_dim=2, activation='relu'))
model.add(Dense(1,activation='sigmoid'))

model.compile(loss="binary_crossentropy", optimizer='adam', metrics=['binary_accuracy'])

model.fit(training_data,target_data, epochs=1000, verbose=2)

print(model.predict(training_data))
print("\n Accuracy : %.4f" % (model.evaluate(training_data,target_data)[1]))


```

```python
# seed 값 생성
seed = 0
np.random.seed(seed)
tf.random.set_seed(seed)

#데이터 로드
dataset = np.loadtxt("./dataset/pima-indians-diabetes.csv", delimiter=",")
X = dataset[:,0:8]
Y = dataset[:,8]

model = Sequential()
model.add(Dense(200,input_dim=8, activation='relu'))
model.add(Dense(8,activation='relu'))
model.add(Dense(8,activation='relu'))
model.add(Dense(8,activation='relu'))
model.add(Dense(1,activation='sigmoid'))

#모델 컴파일
model.compile(loss='binary_crossentropy',optimizer='adam', metrics=['accuracy'])

#모델 실행
model.fit(X,Y,epochs=800, batch_size=10)

#결과 출력
print("\n Accuract:%.4f" % (model.evaluate(X,Y)[1]))
```

## 다중 분류 문제

```python
df = pd.read_csv('./dataset/iris.csv', names=['sepal_length','sepal_width','petal_length','petal_width','species'])
df.head()
```

```python
sns.pairplot(df,hue='species')
plt.show()
```

```python
seed =0
np.random.seed(seed)
tf.random.set_seed(seed)

df = pd.read_csv('./dataset/iris.csv', names=['sepal_length','sepal_width', 'petal_length','petal_width','species'])


#데이터 분류
dataset = df.values
X = dataset[:,0:4].astype(float)
Y_obj = dataset[:,4]

#문자열을 숫자로 변환
e = LabelEncoder()
e.fit(Y_obj)
Y = e.transform(Y_obj)
Y_encoded = np_utils.to_categorical(Y)

#모델의 설정
model = Sequential()
model.add(Dense(16,input_dim=4, activation='relu'))
model.add(Dense(3, activation='softmax')) #다중분류를 위한 softmax사용

#모델 컴파일
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

#모델 실행
model.fit(X,Y_encoded, epochs=50, batch_size=1)

#결과 출력
print("\nAccuracy: %.4f" % (model.evaluate(X,Y_encoded)[1]))
```
