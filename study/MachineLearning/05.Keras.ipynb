```python
from keras.models import Sequential
from keras.layers import Dense
from keras import optimizers
import numpy as np
import tensorflow as tf
```

```python
seed =0
np.random.seed(seed)
tf.random.set_seed(seed)

Data_set = np.loadtxt('./dataset/ThoraricSurgery.csv',delimiter=",")
X = Data_set[:,0:17]
Y = Data_set[:,17]

#딥러닝 구조를 결정
model = Sequential()
model.add(Dense(10,input_dim=17, activation='relu')) # 이 층에서 노드를 몇개 만들지 결정
model.add(Dense(1,activation='sigmoid'))   #출력층

#딥러닝을 실행합니다.
model.compile(loss='mean_squared_error',optimizer='adam', metrics=['accuracy'])
model.fit(X,Y,epochs=30,batch_size=10) # ==session.run() , epochs:반복수 , batch_size:샘플 한번에 몇개 처리?


#결과 출력
print("\nAccuracy:%.4f" % model.evaluate(X,Y)[1]) # 정확도 평가
```

## XOR를 케라스로 해보기

```python
training_data = np.array([[0,0],[0,1],[1,0],[1,1]], "float32")
target_data = np.array([[0],[1],[1],[0]],"float32")

model = Sequential()
model.add(Dense(5,input_dim=2, activation='relu'))
model.add(Dense(1,activation='sigmoid'))

model.compile(loss="binary_crossentropy", optimizer='adam', metrics=['binary_accuracy'])

model.fit(training_data,target_data, epochs=1000, verbose=2)

print(model.predict(training_data))
print("\n Accuracy : %.4f" % (model.evaluate(training_data,target_data)[1]))


```

## 당뇨 데이터 실습

```python
Data_set = np.loadtxt('./dataset/data-03-diabetes.csv',delimiter=",")
X = Data_set[:,0:7]
Y = Data_set[:,8]


model = Sequential()
model.add(Dense(30,input_dim=7, activation='relu')) 
model.add(Dense(1,activation='sigmoid')) 

#딥러닝을 실행합니다.
model.compile(loss='mean_squared_error',optimizer='adam', metrics=['accuracy'])
model.fit(X,Y,epochs=500,batch_size=2) 


#결과 출력
print("\nAccuracy:%.4f" % model.evaluate(X,Y)[1]) # 정확도 평가


```
